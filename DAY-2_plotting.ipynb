```python
#there are five subjects psychology,nano technology,medicine,oncology,anthropology
#two students have applied to do masters in any of these five fields with a line plot show in which field they can exel . the cut off for each subject is 70
#now student1=[67,45,73,71,57]      student2=[43,67,68,53,85]
a=["psychology","nanotechnology","medicine ","oncology","anthropology"]
x=[67,45,73,71,57]
y=[43,67,68,53,85]
b=[70,70,70,70,70]
p.scatter(a,x,label="student 1",color='k',marker='p')  #student 1
p.scatter(a,y,label="student 2",color='b',marker='p')  #student 2
p.plot(a,b,label="cutoff",color='r',linestyle=":",linewidth=4)  #cutoff
p.xlabel("Subjects")
p.ylabel("Marks")
p.legend()
```

```python
#An india vs australia is on live . india has given 298 as target for australia ,
#the runs for every five overs is given as follows 25,51,84,131,160,189,220,250,267,297
#Australia have completed 25 overs 15,41,94,110,151,
#plot overs vs runs for both the countries with india has a blue line and australia has a yellow line with required labels and legend
#the marker for india is flag and marker for australia is plus .
#take overs from a range function create runs as arrays.linewidth=2

import numpy as n
from matplotlib import pyplot as p
p.xlabel("OVERS")
p.ylabel("RUNS")
a=n.array([25,51,84,131,160,189,220,250,267,297])
b=n.array([15,41,94,110,151])
c=n.arange(5,51,5)
d=n.arange(5,26,5)
p.title("INDIA  v/s  AUSTRALIA")
p.plot(c,a,label="India",color='b',marker=">",linewidth=2)
p.plot(d,b,label="Australia",color='y',marker='p',linewidth=2)
p.legend()
p.grid()
```
