```python
A= n.array([34,56,78,90]).reshape(2,2)
print(A)
print("----------------")
print(A.max(axis=0))    #axis=0  -column
print("----------------")
print(A.min(axis=1))    #axis=1  -row
```

```python
#joining arrays
A = n.array([34,56,78,90]).reshape(2,2)
B = n.array([7,9,3,12]).reshape(2,2)
print(A)
print("-------------")
print(B)
print("-------------")
print("After vertically stacking the arrays")
print(n.stack((A,B),axis=0))
print("--------------")
print(n.stack((A,B),axis=1))
```

```python
A=n.arange(30).reshape(2,3,5)  #(g,r,c)
print(A)
print("After dstack")
print(n.dstack(A))  #no of rows --> no of groups
#no of col in every group --> no.of rows in every group
```
